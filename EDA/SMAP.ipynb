```python
import pickle
import os
import matplotlib.pyplot as plt
from utils.tools import plot_anomaly
```

## Load

```python
base_dir = "./data/SMAP/"
os.listdir(base_dir)
```

```python
with open(os.path.join(base_dir, "SMAP_train.pkl"), 'rb') as f:
    SMAP_train = pickle.load(f)
with open(os.path.join(base_dir, "SMAP_test.pkl"), 'rb') as f:
    SMAP_test = pickle.load(f)
with open(os.path.join(base_dir, "SMAP_test_label.pkl"), 'rb') as f:
    SMAP_test_label = pickle.load(f)
```

```python
SMAP_train.shape, SMAP_test.shape, SMAP_test_label.shape
```

## Visualize

```python
L, C = SMAP_train.shape
for c in range(C):
    print(f"channel {c}")
    plt.figure(figsize=(20, 6))
    plt.plot(SMAP_train[:,c])
    plt.show()

    plt.figure(figsize=(20, 6))
    plt.plot(SMAP_test[:,c])
    plot_anomaly(plt, SMAP_test_label)
    plt.show()
```
