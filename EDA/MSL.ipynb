```python
import pickle
import os
import matplotlib.pyplot as plt
from utils.tools import plot_anomaly
```

## Load

```python
base_dir = "./data/MSL/"
os.listdir(base_dir)
```

```python
with open(os.path.join(base_dir, "MSL_train.pkl"), 'rb') as f:
    MSL_train = pickle.load(f)
with open(os.path.join(base_dir, "MSL_test.pkl"), 'rb') as f:
    MSL_test = pickle.load(f)
with open(os.path.join(base_dir, "MSL_test_label.pkl"), 'rb') as f:
    MSL_test_label = pickle.load(f)
```

```python
MSL_train.shape, MSL_test.shape, MSL_test_label.shape
```

```python
MSL_test_label
```

## Visualize

```python
L, C = MSL_train.shape
for c in range(C):
    print(f"channel {c}")
    plt.figure(figsize=(20, 6))
    plt.plot(MSL_train[:,c])
    plt.show()

    plt.figure(figsize=(20, 6))
    plt.plot(MSL_test[:,c])
    plot_anomaly(plt, MSL_test_label)
    plt.show()
```
